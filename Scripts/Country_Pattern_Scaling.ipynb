Now calculate regression coefficients for each country and model

```python
def regcoefs2d(x,y2d,order):
    fits = np.polyfit(x,y2d,order)
    return(fits)

def calc_predicted_t(c,coefs,order):
    if order == 1:
        predicted = coefs[1,:,:,:] + (c*coefs[0,:,:,:])
    if order == 2:
        predicted = coefs[2,:,:,:] + (c*coefs[1,:,:,:]) + ((c**2) * coefs[0,:,:,:])
    return(predicted)
```

```python
iso = t_diff.coords["iso"]
order = 1 # linear reg
orders = np.arange(0,order+1,1)
coefs = xr.DataArray(np.full((len(orders),len(ensembles),len(members),len(iso)),np.nan),
                             coords=[orders,ensembles,members,iso],
                             dims=["order","ensemble","member","iso"])

for ee in np.arange(0,len(ensembles),1):

    e = ensembles[ee]
    if ee == 0:
        ens_members = models_cmip6
    else:
        ens_members = cesm1_sfle_runs
        
    for mm in np.arange(0,len(ens_members),1):
        x = gmst_diff[ee,mm,:]
        y = t_diff[ee,mm,:,:]
        coefs[:,ee,mm,:] = regcoefs2d(x.values,y,order)
```

```python
em_coefs = coefs.mean(dim="member").mean(dim="ensemble")
em_slope = em_coefs[0,:]
```

Read FAIR simulations and predict temperature

```python
# Number of FAIR runs
n_fair_runs = 250
accounting = "territorial"
y1_shares = 1990
y2_shares = 2014

# read in FAIR GMST
y1_fair = 1850
y2_fair = 2014
fair_gmst = xr.open_dataset(loc_gmst_fair+"FAIR_GMST_PPE"+str(n_fair_runs)+"_"+accounting+"_shares"+str(y1_shares)+"-"+str(y2_shares)+"_"+str(y1_fair)+"-"+str(y2_fair)+".nc")
gmst_fair_all = fair_gmst.t_global
gmst_fair_subtract = fair_gmst.t_subtract
gmst_fair_pic = fair_gmst.t_pic
iso_attr = fair_gmst.coords["iso_attr"]
fair_runs = fair_gmst.coords["fair_run"]
gmst_diff_fair_all = gmst_fair_all - gmst_fair_pic
gmst_diff_fair_subtract = gmst_fair_subtract - gmst_fair_pic
```

Write out temp difference and stuff from historical FAIR simulations

```python
t_hist_pred = calc_predicted_t(gmst_diff_fair_all,coefs,order)
```

```python
t_hist_pred_em = t_hist_pred.mean(dim=["ensemble","member","fair_run"])
qs = [0.025,0.975]
t_hist_pred_pctiles = t_hist_pred.quantile(qs,dim=["ensemble","member","fair_run"]) 
```

Write out

```python
y1_out = np.amin(t_hist_pred_em.time.values)
y2_out = np.amax(t_hist_pred_em.time.values)

fair_temp_ds = xr.Dataset({"t_predicted_mean":(["iso","time"],t_hist_pred_em),
                           "t_predicted_pctiles":(["quantile","iso","time"],t_hist_pred_pctiles)},
                          coords={"iso":(["iso"],t_hist_pred_em.iso.values),
                                  "time":(["time"],t_hist_pred_em.time.values),
                                  "quantile":(["quantile"],qs)})

fair_temp_ds.attrs["creation_date"] = str(datetime.datetime.now())
fair_temp_ds.attrs["created_by"] = "Christopher Callahan, Christopher.W.Callahan.GR@dartmouth.edu"
fair_temp_ds.attrs["variable_description"] = "predicted historical temperatures from FaIR pattern scaling"
fair_temp_ds.attrs["created_from"] = os.getcwd()+"/Country_Pattern_Scaling.ipynb"
fair_temp_ds.attrs["ensemble_details"] = "Perturbed-parameter ensemble for carbon cycle and TCR/ECS"
fair_temp_ds.attrs["emissions_subtraction"] = "Leave-one-out country attribution using shares of CEDS emissions"
fair_temp_ds.attrs["emissions_accounting"] = accounting+"-based emissions accounting"

fname_out = loc_out_tdiff_global+"FaIR_country_temp_scaling_prediction_"+str(y1_out)+"-"+str(y2_out)+".nc"
fair_temp_ds.to_netcdf(fname_out,mode="w")
print(fname_out)
```
