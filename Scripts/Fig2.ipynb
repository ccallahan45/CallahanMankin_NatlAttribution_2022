Map helper function

```python
def add_country_geometries(input_vals,iso_vals,iso_plot,shapefile,shapefile_iso_name,cbar):

    for i in iso_plot:
        if i in iso_vals:
            val = input_vals.loc[i].values
            if (~np.isnan(val)):
                ax.add_geometries(shapefile.loc[shapefile[shapefile_iso_name].values==i,"geometry"],ccrs.PlateCarree(),
                                  facecolor=cbar.to_rgba(val),edgecolor="black",linewidth=0.5)
            else:
                ax.add_geometries(shapefile.loc[shapefile[shapefile_iso_name].values==i,"geometry"],ccrs.PlateCarree(),
                                  facecolor=[0.95,0.95,0.95],edgecolor="black",linewidth=0.5)
        else:
            ax.add_geometries(shapefile.loc[shapefile[shapefile_iso_name].values==i,"geometry"],ccrs.PlateCarree(),
                              facecolor=[0.95,0.95,0.95],edgecolor="black",linewidth=0.5)
```

Get US-attributable damages for each country

```python
country_to_attr = "USA"
fname_str = damagefunc+"_"+accounting+"_shares"+str(y1_shares)+"-"+str(y2_shares)+"_"+str(y1_damages)+"-"+str(y2_damages)
files_list = [x for x in os.listdir(loc_damages_country+country_to_attr+"/") if ("global" not in x)&(fname_str in x)&("fulldist" not in x)&("ttest" not in x)]
isos_attr = [x[15:18] for x in files_list]

for i in isos_attr:
    fname = country_to_attr+"-attributed_"+i+"_gdp_damages_"+fname_str+".nc"
    damage_to_country = xr.open_dataset(loc_damages_country+country_to_attr+"/"+fname)
    if i == isos_attr[0]:
        damages_mean = damage_to_country.pct_damage_mean
        damages_sd = damage_to_country.pct_damage_sd_total
        damages_sd_boot = damage_to_country.pct_damage_sd_boot
        damages_sd_fair = damage_to_country.pct_damage_sd_fair
        damages_sd_iv = damage_to_country.pct_damage_sd_iv
        damages_sd_mdl = damage_to_country.pct_damage_sd_mdl
    else:
        damages_mean = xr.concat([damages_mean,damage_to_country.pct_damage_mean],dim="iso")
        damages_sd = xr.concat([damages_sd,damage_to_country.pct_damage_sd_total],dim="iso")
        damages_sd_boot = xr.concat([damages_sd_boot,damage_to_country.pct_damage_sd_boot],dim="iso")
        damages_sd_fair = xr.concat([damages_sd_fair,damage_to_country.pct_damage_sd_fair],dim="iso")
        damages_sd_iv = xr.concat([damages_sd_iv,damage_to_country.pct_damage_sd_iv],dim="iso")
        damages_sd_mdl = xr.concat([damages_sd_mdl,damage_to_country.pct_damage_sd_mdl],dim="iso")
damages_mean.coords["iso"] = isos_attr
damages_sd.coords["iso"] = isos_attr
damages_sd_boot.coords["iso"] = isos_attr
damages_sd_fair.coords["iso"] = isos_attr
damages_sd_iv.coords["iso"] = isos_attr
damages_sd_mdl.coords["iso"] = isos_attr
```

```python
fig = plt.figure(figsize=(11,15))
rcParams["font.family"] = "sans-serif"
rcParams["font.sans-serif"] = ["Helvetica Neue"]
rcParams["font.size"] = 20.0
rcParams["axes.linewidth"] = 1
rcParams['xtick.major.size'] = 4
rcParams['xtick.major.width'] = 1
rcParams['ytick.major.size'] = 4
rcParams['ytick.major.width'] = 1

tfs = 22

gs1 = gridspec.GridSpec(1,1)
gs1.update(left=0.1,right=0.96,top=0.9,bottom=0.58,wspace=0.03,hspace=0.2)

ax = plt.subplot(gs1[0,0],projection=ccrs.Robinson())

ax.set_extent([-180,180,-63,90],crs=ccrs.PlateCarree())
ax.coastlines(linewidth=0.5)
ax.add_feature(cart.feature.OCEAN, edgecolor="white",facecolor=[0.93,0.93,0.93])


cmin = -2
cmax = 2
nrange = 17
bounds = np.linspace(cmin,cmax,nrange)
cols = plt.get_cmap('RdYlBu')(np.linspace(0,1,len(bounds)+1))
cmap = colors.ListedColormap(cols[1:-1])
cmap.set_over(cols[-1])
cmap.set_under(cols[0])
norm = colors.BoundaryNorm(boundaries=bounds, ncolors=len(bounds)-1)
cbar = plt.cm.ScalarMappable(norm=norm,cmap=cmap)


for i in damages_mean.iso.values:
    if i in iso_shp:
        val = damages_mean.loc[i,2014].values
        p_val = 0 #p.loc[i].values
        if ((~np.isnan(val)) & (p_val < 0.05)):
                ax.add_geometries(shp.loc[shp["ISO3"].values==i,"geometry"],ccrs.PlateCarree(),
                                  facecolor=cbar.to_rgba(val),edgecolor="black",linewidth=0.5)
        elif ((~np.isnan(val)) & (p_val >= 0.05)):
            ax.add_geometries(shp.loc[shp["ISO3"].values==i,"geometry"],ccrs.PlateCarree(),
                                  facecolor=cbar.to_rgba(val),edgecolor="black",linewidth=0.5,
                             hatch="///")
        else:
            ax.add_geometries(shp.loc[shp["ISO3"].values==i,"geometry"],ccrs.PlateCarree(),
                                  facecolor=[0.95,0.95,0.95],edgecolor="black",linewidth=0.5)
    else:
        ax.add_geometries(shp.loc[shp["ISO3"].values==i,"geometry"],ccrs.PlateCarree(),
                              facecolor=[0.95,0.95,0.95],edgecolor="black",linewidth=0.5)

if damagefunc=="BHMSR":
    funcstr = "Short-run dmg."
st = str(y1_damages)+"-"+str(y2_damages)+"\n"+accounting.title()+"\n"+funcstr
plt.text(0.03,0.25,st,ha="left",fontsize=18,
         transform=ax.transAxes)
        
plt.title("U.S.-attributable damages",pad=10,fontsize=tfs)

cb = plt.colorbar(cbar,orientation='horizontal',
                  drawedges=True,extend="both",
                  shrink=0.72,pad=0.02,
                  ticks=[-3,-2,-1,0,1,2,3],aspect=23)
cb.ax.tick_params(direction="in",width=0.05)
cb.set_label("GDP per capita change (%)")


#bgbox = ax.inset_axes([0.025, 0.235, 0.19, 0.16])
#[bgbox.spines[k].set_visible(False) for k in bgbox.spines]
#bgbox.tick_params(axis = 'both',left = False,top = False,right = False,bottom = False,
#                  labelleft = False,labeltop = False,labelright = False,labelbottom = False)
#bgbox.set_facecolor([0.7,0.7,0.7,0.4])

# likely: 0.95 (66%)
# very likely: 1.64 (90%)
# virtually certain: 2.57 (99%)
# marvel et al., nature, 2019
#snr_pct1 = np.around((np.sum(snr.values>0.95)/len(snr))*100)
#snr_pct2 = np.around((np.sum(snr.values>1.64)/len(snr))*100)
#plt.figtext(0.13,0.38,"Likely: "+str(snr_pct1)[:2]+"%",
#           fontsize=16)
#plt.figtext(0.13,0.35,"Very likely: "+str(snr_pct2)[:2]+"%",
#           fontsize=16)



rcParams["axes.linewidth"] = 1.3
rcParams['xtick.major.size'] = 8
rcParams['xtick.major.width'] = 1.3
rcParams['ytick.major.size'] = 8
rcParams['ytick.major.width'] = 1.3

gs2 = gridspec.GridSpec(1,2)
gs2.update(left=0.14,right=0.9,top=0.51,bottom=0.2,wspace=0.22,hspace=0.2)

damages_mean.name = "gdp_damage"
full_data = damages_mean[:,-1].to_dataframe()
full_data_sorted = full_data.sort_values(by="gdp_damage").dropna()
rank_num = 5

data_sorted_bottom = full_data_sorted[0:rank_num]
data_sorted_top = full_data_sorted[full_data_sorted.shape[0]-rank_num:(full_data_sorted.shape[0])]

carbon_color = "olivedrab"
boot_color = "darkturquoise"
iv_color = "cornflowerblue"
ms_color = "salmon"

nms_bottom = ["UAE","Mauritania","Saudi Arabia",\
             "Oman","Iraq"]
nms_top = ["Iceland","Russia",\
          "Finland","Mongolia","Greenland"]


for k in np.arange(0,2,1):
    
    ax = plt.subplot(gs2[0,k])
    
    if k == 0:
        plt.axhline(y=0,linewidth=1.3,linestyle="--",color="black")
        data_sorted = data_sorted_bottom    
        for j in np.arange(0,len(data_sorted.index.values),1):
            i = data_sorted.index.values[j]
            mn = data_sorted.loc[i,"gdp_damage"]
            #td_iso = tdiff_attr_em.loc[i,2014]
            #damage_per_temp = np.around(mn/td_iso.values)
            
            var_carbon = damages_sd_fair.loc[i,2014].values**2
            var_boot = damages_sd_boot.loc[i,2014].values**2
            var_iv = damages_sd_iv.loc[i,2014].values**2
            var_model = damages_sd_mdl.loc[i,2014].values**2

            p0, = plt.plot([j-0.3,j+0.3],[mn,mn],linewidth=3,color="black",zorder=100)

            # 90% CI = 1.654, 95% CI = 1.96
            # see Lehner et al., Earth System Dynamics, 2020 
            # for more on the following calculation
            ci_z = 1.96

            Fnum = (np.sqrt(var_carbon) + np.sqrt(var_boot) + np.sqrt(var_iv) + np.sqrt(var_model))
            Fdenom = np.sqrt(var_carbon + var_boot + var_iv + var_model)
            F = Fnum/Fdenom
            ci_carbon = (ci_z * np.sqrt(var_carbon))/F
            ci_boot = (ci_z * np.sqrt(var_boot))/F
            ci_model = (ci_z * np.sqrt(var_model))/F
            ci_iv = (ci_z * np.sqrt(var_iv))/F

            ci_lower = mn - (ci_carbon + ci_boot + ci_model + ci_iv)
            p1 = ax.fill_between([j-0.25,j+0.25],ci_lower,ci_lower+(2*ci_carbon),
                                facecolor=carbon_color,edgecolor="black",linewidth=1)
            p2 = ax.fill_between([j-0.25,j+0.25],ci_lower+(2*ci_carbon),
                                 ci_lower+(2*ci_carbon)+(2*ci_boot),
                                facecolor=boot_color,edgecolor="black",linewidth=1)
            p3 = ax.fill_between([j-0.25,j+0.25],ci_lower+(2*ci_carbon)+(2*ci_boot),
                                 ci_lower+(2*ci_carbon)+(2*ci_boot)+(2*ci_model),
                                facecolor=ms_color,edgecolor="black",linewidth=1)
            p4 = ax.fill_between([j-0.25,j+0.25],ci_lower+(2*ci_carbon)+(2*ci_boot)+(2*ci_model),
                                 ci_lower+(2*ci_carbon)+(2*ci_boot)+(2*ci_model)+(2*ci_iv),
                                facecolor=iv_color,edgecolor="black",linewidth=1)
            #if i == "ARE":
            #    damage_temp_str = str(int(damage_per_temp))+" "+r"$\frac{\%}{\mathrm{K}}$"
            #else:
            #    damage_temp_str = str(int(damage_per_temp))
            #plt.text(j-0.2,-11.6,damage_temp_str,
            #        fontsize=14)
            
        #plt.xticks(ticks=[0,1,2,3,4],labels=data_sorted.index,
        #      rotation=0)
        plt.xticks(ticks=[0,1,2,3,4],labels=nms_bottom,
              rotation=45,ha="right",rotation_mode="anchor")
        #plt.ylim([-12,8])
        
    else:
        
        data_sorted = data_sorted_top
        
        for j in np.arange(0,len(data_sorted.index.values)-1,1):
            i = data_sorted.index.values[j]
            
            mn = data_sorted.loc[i,"gdp_damage"]
            #td_iso = tdiff_attr_em.loc[i,2014]
            #damage_per_temp = np.around(mn/td_iso.values)
            
            var_carbon = damages_sd_fair.loc[i,2014].values**2
            var_boot = damages_sd_boot.loc[i,2014].values**2
            var_iv = damages_sd_iv.loc[i,2014].values**2
            var_model = damages_sd_mdl.loc[i,2014].values**2

            p0, = plt.plot([j-0.3,j+0.3],[mn,mn],linewidth=3,color="black",zorder=100)

            # 90% CI = 1.654, 95% CI = 1.96
            # see Lehner et al., Earth System Dynamics, 2020 
            # for more on the following calculation
            ci_z = 1.96

            Fnum = (np.sqrt(var_carbon) + np.sqrt(var_boot) + np.sqrt(var_iv) + np.sqrt(var_model))
            Fdenom = np.sqrt(var_carbon + var_boot + var_iv + var_model)
            F = Fnum/Fdenom
            ci_carbon = (ci_z * np.sqrt(var_carbon))/F
            ci_boot = (ci_z * np.sqrt(var_boot))/F
            ci_model = (ci_z * np.sqrt(var_model))/F
            ci_iv = (ci_z * np.sqrt(var_iv))/F

            ci_lower = mn - (ci_carbon + ci_boot + ci_model + ci_iv)
            p1 = ax.fill_between([j-0.25,j+0.25],ci_lower,ci_lower+(2*ci_carbon),
                                facecolor=carbon_color,edgecolor="black",linewidth=1)
            p2 = ax.fill_between([j-0.25,j+0.25],ci_lower+(2*ci_carbon),
                                 ci_lower+(2*ci_carbon)+(2*ci_boot),
                                facecolor=boot_color,edgecolor="black",linewidth=1)
            p3 = ax.fill_between([j-0.25,j+0.25],ci_lower+(2*ci_carbon)+(2*ci_boot),
                                 ci_lower+(2*ci_carbon)+(2*ci_boot)+(2*ci_model),
                                facecolor=ms_color,edgecolor="black",linewidth=1)
            p4 = ax.fill_between([j-0.25,j+0.25],ci_lower+(2*ci_carbon)+(2*ci_boot)+(2*ci_model),
                                 ci_lower+(2*ci_carbon)+(2*ci_boot)+(2*ci_model)+(2*ci_iv),
                                facecolor=iv_color,edgecolor="black",linewidth=1)
            
            #plt.text(j-0.2,-37.8,str(int(damage_per_temp)),
            #        fontsize=14)
            
        #plt.axhline(y=0,linewidth=1,linestyle="--",color="black")
        plt.ylim([-15,30])
        #plt.yticks([-40,-20,0,20,40,60])
        plt.xticks(ticks=[0,1,2,3,4],labels=nms_top,
              rotation=45,color="black",ha="right",rotation_mode="anchor")
        #ax.tick_params(axis='x', colors=["black","black","black","forestgreen","forestgreen"])
        tickcols = ["black","black","black","black","forestgreen"]
        for ticklabel, tickcolor in zip(plt.gca().get_xticklabels(), tickcols):
            ticklabel.set_color(tickcolor)
        plt.plot([-1,3.5],[0,0],linewidth=1.3,color="black",linestyle="--")
        
        ax2 = ax.twinx()
        
        plt.plot([3.5,5],[0,0],linewidth=1.3,color="forestgreen",linestyle="--")
        
        #plt.xticks(ticks=[3,4],labels=data_sorted.index[3:5],
        #      rotation=0,color="forestgreen")
        plt.ylim([-30,60])
        #plt.yticks(ticks=[-3000,-1500,0,1500,3000],
        #           labels=[r"$-$3e3",r"$-$1e3","0","1e3","3e3"],
        #           color="forestgreen")
        #plt.yticks(ticks=[-15000,-10000,-5000,0,5000,10000,15000],
        #           labels=[r"$-$1.5e4","-1e4","-5e3","0","5e3","1e4","1.5e4"],
        #           color="forestgreen")
        
        
        for j in np.arange(len(data_sorted.index.values)-1,len(data_sorted.index.values),1):
            i = data_sorted.index.values[j]
            
            mn = data_sorted.loc[i,"gdp_damage"]
            #td_iso = tdiff_attr_em.loc[i,2014]
            #damage_per_temp = np.around(mn/td_iso.values)
            
            var_carbon = damages_sd_fair.loc[i,2014].values**2
            var_boot = damages_sd_boot.loc[i,2014].values**2
            var_iv = damages_sd_iv.loc[i,2014].values**2
            var_model = damages_sd_mdl.loc[i,2014].values**2

            p0, = plt.plot([j-0.3,j+0.3],[mn,mn],linewidth=3,color="black",zorder=100)

            # 90% CI = 1.654, 95% CI = 1.96
            # see Lehner et al., Earth System Dynamics, 2020 
            # for more on the following calculation
            ci_z = 1.96

            Fnum = (np.sqrt(var_carbon) + np.sqrt(var_boot) + np.sqrt(var_iv) + np.sqrt(var_model))
            Fdenom = np.sqrt(var_carbon + var_boot + var_iv + var_model)
            F = Fnum/Fdenom
            ci_carbon = (ci_z * np.sqrt(var_carbon))/F
            ci_boot = (ci_z * np.sqrt(var_boot))/F
            ci_model = (ci_z * np.sqrt(var_model))/F
            ci_iv = (ci_z * np.sqrt(var_iv))/F

            ci_lower = mn - (ci_carbon + ci_boot + ci_model + ci_iv)
            p1 = ax2.fill_between([j-0.25,j+0.25],ci_lower,ci_lower+(2*ci_carbon),
                                facecolor=carbon_color,edgecolor="black",linewidth=1)
            p2 = ax2.fill_between([j-0.25,j+0.25],ci_lower+(2*ci_carbon),
                                 ci_lower+(2*ci_carbon)+(2*ci_boot),
                                facecolor=boot_color,edgecolor="black",linewidth=1)
            p3 = ax2.fill_between([j-0.25,j+0.25],ci_lower+(2*ci_carbon)+(2*ci_boot),
                                 ci_lower+(2*ci_carbon)+(2*ci_boot)+(2*ci_model),
                                facecolor=ms_color,edgecolor="black",linewidth=1)
            p4 = ax2.fill_between([j-0.25,j+0.25],ci_lower+(2*ci_carbon)+(2*ci_boot)+(2*ci_model),
                                 ci_lower+(2*ci_carbon)+(2*ci_boot)+(2*ci_model)+(2*ci_iv),
                                facecolor=iv_color,edgecolor="black",linewidth=1)
            
            #plt.text(j-0.2,-3050,str(int(damage_per_temp)),
            #        fontsize=14)
            
    plt.xlim([-0.75,4.75])
    if k == 0:
        plt.ylim([-5,2])
        #plt.yticks([-5,3])
        plt.ylabel("GDP per capita change (%)")
        plt.title("Bottom five",pad=10,fontsize=tfs)
    else:
        
        plt.axvline(x=3.5,linewidth=1.3,color="gray",linestyle="-")
        plt.title("Top five",fontsize=tfs,pad=10)
        
    if k == 0:
        leg = plt.legend([p1,p2,p0,p3,p4],
                   ["Carbon cycle","Dmg. function","Mean","GCM structure","Internal var."],
                   loc="upper left",frameon=False,fontsize=14,ncol=2,
                  handletextpad=0.7,handlelength=1.5,columnspacing=0.7)



plt.figtext(0.13,0.92,r'$\bf{a}$',fontsize=26)
plt.figtext(0.13,0.535,r'$\bf{b}$',fontsize=26)
plt.figtext(0.55,0.535,r'$\bf{c}$',fontsize=26)

plt.savefig("../Figures/Fig2.pdf")

plt.show()
```
